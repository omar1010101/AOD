# Benchmark datasets exploration

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

## Artificial No Anomaly

### Art daily no noise

```python
df1 = pd.read_csv('https://raw.githubusercontent.com/numenta/NAB/master/data/artificialNoAnomaly/art_daily_no_noise.csv')
```

```python
df1.head()
```

```python
df1['timestamp'] = pd.to_datetime(df1['timestamp'])
```

```python
fig, ax = plt.subplots(figsize = (20, 5))
plt.scatter(x=df1['timestamp'], y=df1['value'], s=2);
```

```python
fig, ax = plt.subplots(figsize = (20, 5))
plt.plot(df1['timestamp'], df1['value']);
```
